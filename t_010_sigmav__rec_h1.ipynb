```python
import numpy as np
```

# $\alpha_{\text{rec}}^{\text{rad}}(n{l}) = A_{n{l}} \times 10^{-14} \left( \frac{E_n^{\text{ion}}}{\text{Ry}} \right)^{1/2} \frac{\beta_n^{3/2}}{\beta_n + \chi_{n{l}}}10^{-6} \quad \left[\frac{\text{m}^3}{\text{s}}\right],$
# $\beta_n = \frac{E_n^{\text{ion}}}{nT}$
$A_{nl=1s} = 3.92$\
$\chi_{nl=1s} = 0.35$

```python
def sigmav_rec_h1s(Te):
    Te = np.asarray(Te)
    Te_clip = np.clip(Te,0.1,2.01e4)
    # Information: Janev 1987, p. 32
    n  = 1
    Ry = 13.58
    E_ion_n = Ry/n
    A_nl = 3.92
    X_nl = 0.35
    B_n = E_ion_n/Te_clip
    result = (A_nl*(1e-14))*np.sqrt(E_ion_n/Ry)*((B_n**1.5)/(B_n+X_nl))*1e-6
    return result
```

```python
import matplotlib.pyplot as plt
Te = np.logspace(-1,5,100)
Te_clip = np.clip(Te,0.1,2.01e4)
# Te_clip
result = sigmav_rec_h1s(Te)
plt.plot(Te_clip,result)
plt.yscale('log')
plt.xscale('log')
plt.grid(True,'both')
plt.show()
```
