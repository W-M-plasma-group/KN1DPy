```python
class KineticHShared:
    def __init__(self):
        # Variables compartidas inicializadas como None
        self.vx_s = None
        self.vr_s = None
        self.x_s = None
        self.Tnorm_s = None
        self.mu_s = None

# Instancia global creada UNA SOLA VEZ
shared_data = KineticHShared()

def kinetic_h(vx, vr, x, Tnorm, mu):
    global shared_data  # Accedemos a la instancia global

    # Inicializar New_Grid
    New_Grid = 1
    print(shared_data.vx_s)

    # Verificar si vx_s está definida (no es None)
    if shared_data.vx_s is not None:
        print('shared_data.vx_s Sí está definida')
        test = 0

        # Comparar vx_s con vx
        test += sum(1 for a, b in zip(shared_data.vx_s, vx) if a != b)
        # Comparar vr_s con vr
        test += sum(1 for a, b in zip(shared_data.vr_s, vr) if a != b)
        # Comparar x_s con x
        test += sum(1 for a, b in zip(shared_data.x_s, x) if a != b)
        # Comparar Tnorm_s con Tnorm
        test += sum(1 for a, b in zip(shared_data.Tnorm_s, Tnorm) if a != b)
        # Comparar mu_s con mu
        test += sum(1 for a, b in zip(shared_data.mu_s, mu) if a != b)

        # Si no hay diferencias, establecer New_Grid = 0
        if test <= 0:
            New_Grid = 0

    # Actualizar las variables compartidas con los valores actuales
    shared_data.vx_s = vx
    shared_data.vr_s = vr
    shared_data.x_s = x
    shared_data.Tnorm_s = Tnorm
    shared_data.mu_s = mu

    return New_Grid

for i in range(5):
    vx = [i, i + 1, i + 2]
    vr = [i, i + 1, i + 2]
    x = [i, i + 1, i + 2]
    Tnorm = [i, i + 1, i + 2]
    mu = [i, i + 1, i + 2]

for i in range(5):
    result = kinetic_h(vx, vr, x, Tnorm, mu)
    print(f"Iteración {i}: New_Grid = {result}")
    print(f'shared_data.vx_s: {shared_data.vx_s}')
    print(f'vx: {vx}')
    print(35*'$')
```

```python
# import numpy as np
# import matplotlib.pyplot as plt
# from PIL import Image

# # Cargar imagen y asegurar formato RGB
# ruta_imagen = "Arsenal_logo.png"
# imagen = Image.open(ruta_imagen).convert("RGB")

# # Redimensionar imagen a una escala más pequeña para análisis
# # dim = 64  
# imagen_resized = imagen.resize((dim, dim), Image.Resampling.NEAREST)

# # Convertir imagen a matriz NumPy
# imagen_array = np.array(imagen_resized)

# # Obtener colores únicos
# colores_unicos = np.unique(imagen_array.reshape(-1, 3), axis=0)

# # Determinar dimensiones de la cuadrícula
# num_colores = len(colores_unicos)
# cols = 10  # Número de columnas en la cuadrícula
# rows = (num_colores // cols) + 1  # Número de filas necesarias

# # Crear figura
# fig, ax = plt.subplots(rows, cols, figsize=(10, rows))

# # Rellenar la cuadrícula con los colores
# for i, color in enumerate(colores_unicos):
#     fila = i // cols
#     columna = i % cols
#     ax[fila, columna].imshow([[color / 255]])  # Normalizar a [0,1] para Matplotlib
#     ax[fila, columna].axis("off")  # Ocultar ejes

# # Eliminar subgráficos vacíos
# for j in range(i + 1, rows * cols):
#     fila = j // cols
#     columna = j % cols
#     ax[fila, columna].axis("off")

# plt.show()
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Cargar imagen y asegurar formato RGB
ruta_imagen = "Arsenal_logo.png"
ruta_imagen = "City_logo.png"
ruta_imagen = "Inter_logo.png"
ruta_imagen = 'Bayern_logo.png'
imagen = Image.open(ruta_imagen).convert("RGB")

# Redimensionar imagen a una escala más pequeña para análisis
dim = 64  
imagen_resized = imagen.resize((dim, dim), Image.Resampling.NEAREST)

# Convertir imagen a matriz NumPy
imagen_array = np.array(imagen_resized)

# Obtener colores únicos
colores_unicos = np.unique(imagen_array.reshape(-1, 3), axis=0)

# Función de ordenación personalizada
def ordenar_colores(color):
    # Priorizar R descendente, G descendente y B descendente
    return (-color[0], -color[1], -color[2])

# Ordenar los colores según el criterio
colores_ordenados = sorted(colores_unicos, key=ordenar_colores)

# Determinar dimensiones de la cuadrícula
num_colores = len(colores_ordenados)
cols = 10  # Número de columnas en la cuadrícula
rows = (num_colores // cols) + 1  # Número de filas necesarias

# Crear figura
fig, ax = plt.subplots(rows, cols, figsize=(10, rows))

# Rellenar la cuadrícula con los colores ordenados
for i, color in enumerate(colores_ordenados):
    fila = i // cols
    columna = i % cols
    ax[fila, columna].imshow([[color / 255]])  # Normalizar a [0,1] para Matplotlib
    ax[fila, columna].axis("off")  # Ocultar ejes

# Eliminar subgráficos vacíos
for j in range(i + 1, rows * cols):
    fila = j // cols
    columna = j % cols
    ax[fila, columna].axis("off")

plt.tight_layout()
plt.show()

```
